2. Haga la gráfica de la función $f(x)=1 + \frac{x^2}{2!} + \frac{x^4}{4!}$ y de la función $\cosh (x)$ en la misma gráfica. Como se comparan las dos funciones anteriores en el intervalo de $-1$ a $1$. Explique las similitudes y diferencias.

````python
#Importamos librerias
import numpy as np
import matplotlib.pyplot as plt

#Funciones
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n-1)

def f(x):
  return 1 + x**2/factorial(2) + x**4/factorial(4)

#Rango
x_values = np.linspace(-1,1)
y_values = f(x_values)
y_values_c = np.cosh(x_values)

plt.plot(x_values, y_values, label=r'$f(x)=1+\frac{x^2}{2}+\frac{x^4}{2\cdot3\cdot4}$', color ='blue')
plt.plot(x_values, y_values_c, label='$\cosh(x)$', color = 'red', linestyle = '--')
plt.title('Comparación de $f(x)$ y $\cosh(x)$')
plt.legend()
plt.show()
````

4. Realice una función que reciba un número telefónico de 10 dígitos y que a la salida proporcione la suma y el producto de todos de todos sus dígitos distintos de cero. Compruebe el funcionamiento de su función con algún teléfono. Ejemplo:
```
In: sumar_tel(5545973256)
Out: Suma = 51 ; Producto = 5,670,000
```

````python
def sum_tel(num):

#Cadena para cada dígito:
    num_str = str(num)

#Sumando y multiplicando:
    sum = 0
    plus = 1

#Hacemos el ciclo
    for nums in num_str:
      nums_int = int(nums)
#Excluyendo el cero:
      if nums_int !=0:
        sum += nums_int
        plus *= nums_int
    return sum, plus

#Probando:
Número = 5545973256
sum, plus = sum_tel(Número)

#Pedimos el resultado
print("Suma=", sum, "Producto =", plus)
````
